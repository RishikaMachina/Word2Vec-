```python
import gensim
from gensim.models import Word2Vec
```

```python
import pandas as pd

with open("medical_train.dat", "r") as fh:
    lines = fh.readlines() 

labels = []
texts = []

for line in lines:
    splitline = line.split('\t')
    labels.append(splitline[0])
    texts.append(splitline[1])

texts_df = pd.DataFrame(texts)
texts_df.columns = ["text"]
```

```python
def punctuation_stopwords(words):
    cleaned_text = []
    stop_words = stopwords.words('english')+list(string.punctuation)
    for w in words:
        w = w.lower()
        if w not in stop_words:
            cleaned_text.append(w)
        else:
            continue
    return cleaned_text
```

```python
from nltk.corpus import stopwords
import string
from nltk.tokenize import word_tokenize

cleaned_text = []
for text in texts:
    a = punctuation_stopwords(word_tokenize(text))
    cleaned_text.append(a)
```

```python
model = Word2Vec(cleaned_text)
```
